```python
import os
import numpy as np
import pandas as pd
import subprocess

# --- Configuration ---
PDK_LIB_PATH = "sky130_fd_pr/models/sky130.lib.spice" # Example path
TEMP = 27
VDS_RANGE = np.arange(0.1, 1.8, 0.05)
VGS_RANGE = np.arange(0.3, 1.8, 0.05)
L_RANGE = [0.15, 0.5, 1.0, 2.0] # Lengths in um
W_REF = 5.0 # Reference width for normalization (gm/W)

def create_netlist(vgs, vds, length, width, filename="tb_mos.spice"):
    """Generates a SPICE netlist for a single operating point."""
    netlist = f"""
    * Characterization Testbench
    .lib "{PDK_LIB_PATH}" tt
    .param width={width}
    .param length={length}

    * Device Under Test (NMOS Example)
    XM1 d g 0 0 sky130_fd_pr__nfet_01v8 w={{width}} l={{length}}

    * Sources
    Vds d 0 {vds}
    Vgs g 0 {vgs}

    * Analysis
    .op
    .save all @m.xm1.msky130_fd_pr__nfet_01v8[gm] @m.xm1.msky130_fd_pr__nfet_01v8[gds] @m.xm1.msky130_fd_pr__nfet_01v8[id]
    .end
    """
    with open(filename, "w") as f:
        f.write(netlist)

def run_simulation(netlist_file):
    """Runs Ngspice and parses the OP output."""
    # Run Ngspice in batch mode (-b)
    cmd = ["ngspice", "-b", netlist_file]
    result = subprocess.run(cmd, capture_output=True, text=True)

    # Simple parsing logic (simulated for this example)
    # In reality, you would parse the .raw file or standard output
    output_data = {}
    for line in result.stdout.split('\n'):
        if "gm" in line: output_data['gm'] = float(line.split('=')[1])
        if "gds" in line: output_data['gds'] = float(line.split('=')[1])
        if "id" in line: output_data['id'] = float(line.split('=')[1])

    return output_data

# --- Main Execution ---
data_list = []

print("Starting LUT Generation...")
for l in L_RANGE:
    for vgs in VGS_RANGE:
        for vds in VDS_RANGE:
            # Create and Run
            create_netlist(vgs, vds, l, W_REF)
            # res = run_simulation("tb_mos.spice")

            # MOCK DATA for demonstration (since we can't run ngspice here)
            # In real usage, use the 'res' from above.
            id_val = 0.5 * 200e-6 * (W_REF/l) * (vgs - 0.3)**2 * (1 + 0.1*vds)
            gm_val = 2 * id_val / (vgs - 0.3)
            gds_val = 0.1 * id_val

            # Store normalized data (independent of W)
            entry = {
                'L': l,
                'Vgs': vgs,
                'Vds': vds,
                'gm_W': gm_val / W_REF,   # Normalized gm
                'gds_W': gds_val / W_REF, # Normalized gds
                'id_W': id_val / W_REF,   # Normalized Current Density
                'gm_id': gm_val / id_val  # gm/Id efficiency
            }
            data_list.append(entry)

# Save to Look-Up Table (CSV or NPY)
df = pd.DataFrame(data_list)
df.to_csv("nmos_lut.csv", index=False)
print("LUT saved to nmos_lut.csv")
```

```python
import pandas as pd
import numpy as np

# 1. Load the LUT (The database generated in Phase 1)
lut = pd.read_csv("nmos_lut.csv")

# 2. Define LDO Specifications (The "Global Exploration" inputs)
TARGET_GAIN = 40.0       # dB
TARGET_BW = 10.0e6       # 10 MHz
CL = 10e-12              # Load Capacitance
IOUT_MAX = 0.01          # 10 mA
V_DROPOUT = 0.2          # 200 mV

print(f"Searching for designs with Gain > {TARGET_GAIN}dB and BW > {TARGET_BW/1e6}MHz...")

# 3. Topology Equations (The "Symbolic Analysis" / MNA)
# Example: Simple Pass Transistor (PMOS) or Common Source Amp
# For this example, we verify if a single stage can meet the specs.
# Gain = gm / gds (Intrinsic)
# BW = gm / (2 * pi * CL) (Dominant pole)

# We can perform vectorized operations on the entire LUT at once
lut['intrinsic_gain'] = lut['gm_W'] / lut['gds_W']
lut['intrinsic_gain_db'] = 20 * np.log10(lut['intrinsic_gain'])

# 4. Filter "Valid" Design Points (The Automation)
# Find rows where intrinsic properties allow meeting the spec
candidates = lut[
    (lut['intrinsic_gain_db'] >= TARGET_GAIN) &
    (lut['Vds'] >= V_DROPOUT)
].copy()

if candidates.empty:
    print("No valid devices found in LUT. Relax specs or expand LUT ranges.")
    exit()

# 5. Sizing (Calculate W)
# We need to size W to handle Iout_max at the specific current density (id_W)
# W_required = Iout_max / id_W
candidates['W_req'] = IOUT_MAX / candidates['id_W']

# Re-calculate exact Gain and BW with the specific W (though gm/gds is W-independent)
# gm_total = gm_W * W_req
candidates['gm_total'] = candidates['gm_W'] * candidates['W_req']
candidates['BW_hz'] = candidates['gm_total'] / (2 * np.pi * CL)

# 6. Final Selection
# Filter candidates that also meet the Bandwidth requirement
valid_designs = candidates[candidates['BW_hz'] >= TARGET_BW]

# Sort by smallest Area (W * L) or Power
valid_designs['Area'] = valid_designs['W_req'] * valid_designs['L']
best_design = valid_designs.sort_values(by='Area').iloc[0]

# --- Output Report ---
print("\n--- Optimal LDO Pass Device Found ---")
print(f"Length (L) : {best_design['L']} um")
print(f"Width (W)  : {best_design['W_req']:.2f} um")
print(f"Bias (Vgs) : {best_design['Vgs']} V")
print(f"Bias (Vds) : {best_design['Vds']} V")
print(f"Result Gain: {best_design['intrinsic_gain_db']:.2f} dB")
print(f"Result BW  : {best_design['BW_hz']/1e6:.2f} MHz")
```

```python
import pandas as pd
import numpy as np
import os

class AnalogSolver:
    def __init__(self, lut_path="master_lut.csv"):
        # Load and clean data
        self.lut = pd.read_csv(lut_path)
        # Filter out cutoff devices to prevent divide-by-zero
        self.lut = self.lut[self.lut['id_W'] > 1e-12].reset_index(drop=True)
        print(f"Loaded LUT with {len(self.lut)} active points.")

    def _filter_device(self, type, v_ds_min, l_min=0):
        """Pre-filters the LUT for specific device types and saturation regions."""
        df = self.lut[
            (self.lut['type'] == type) &
            (self.lut['Vds'] >= v_ds_min) &
            (self.lut['L'] >= l_min)
        ].copy()

        # Calculate Intrinsic Gain (Av0)
        df['Av0'] = df['gm_W'] / (df['gds_W'] + 1e-15)
        df['Av0_db'] = 20 * np.log10(df['Av0'])
        return df

    # ==========================================
    # 1. CSA / CGA / CDA Solver (Saves ALL valid designs)
    # ==========================================
    def solve_single_stage(self, topology, target_gain, cl_load, i_bias_max, filename):
        print(f"\n--- Generating Design Space for {topology} ---")

        # 1. Filter candidates (NMOS in Saturation)
        candidates = self._filter_device("nmos", v_ds_min=0.4)

        # 2. Filter by Gain Spec
        valid = candidates[candidates['Av0_db'] >= target_gain].copy()

        if valid.empty:
            print(f"No designs found for {topology} with Gain > {target_gain}dB")
            return

        # 3. Calculate Topology-Specific Metrics for ALL rows
        # W_req: The width required to support the specific current density given the budget
        # We assume we use the full power budget to maximize speed (Method A)
        # OR we calculate W to hit a specific BW (Method B). Let's use Method A (Fixed Power).

        valid['W_designed'] = i_bias_max / valid['id_W']
        valid['gm_total'] = valid['gm_W'] * valid['W_designed']
        valid['gds_total'] = valid['gds_W'] * valid['W_designed']

        if topology == 'CSA':
            # BW = gm / (2pi * CL)
            valid['BW_Hz'] = valid['gm_total'] / (2 * np.pi * cl_load)
            valid['GBW_Hz'] = valid['BW_Hz'] * valid['Av0']

        elif topology == 'CGA':
            # Input R = 1/gm
            valid['R_in_Ohm'] = 1 / (valid['gm_total'] + 1e-15)

        elif topology == 'CDA':
            # Output R = 1/gm
            valid['R_out_Ohm'] = 1 / (valid['gm_total'] + 1e-15)

        # 4. Save to CSV
        # We select relevant columns to keep the file clean
        cols = ['L', 'Vgs', 'Vds', 'W_designed', 'Av0_db', 'gm_id', 'id_W']
        if topology == 'CSA': cols += ['BW_Hz', 'GBW_Hz']
        if topology == 'CGA': cols += ['R_in_Ohm']
        if topology == 'CDA': cols += ['R_out_Ohm']

        valid[cols].to_csv(filename, index=False)
        print(f"-> Saved {len(valid)} valid {topology} designs to '{filename}'")
        return valid

    # ==========================================
    # 2. Simple OTA Solver (Saves ALL valid designs)
    # ==========================================
    def solve_simple_ota(self, target_gm, filename):
        print(f"\n--- Generating Design Space for 5-T OTA ---")

        # 1. Diff Pair Candidates (NMOS)
        # We need pairs that provide good efficiency
        nmos = self._filter_device("nmos", v_ds_min=0.4, l_min=0.5)

        # 2. Load Candidates (PMOS)
        pmos = self._filter_device("pmos", v_ds_min=0.4, l_min=0.5)

        # CROSS PRODUCT: This is computationally heavy, so we simplify.
        # Instead of every NMOS vs every PMOS, we iterate NMOS and pick the "Best Matching" PMOS.

        results = []

        # Loop through valid NMOS Diff Pairs (Downsample if too slow)
        # We select a subset to keep the file size manageable, or process all.
        for idx, diff in nmos.iterrows():

            # W1 required to hit Target Gm
            W1 = target_gm / diff['gm_W']
            Id_bias_half = diff['id_W'] * W1

            # Find a matching PMOS Load (Same L, capable of handling current)
            # We pick the PMOS with closest L
            matching_loads = pmos[pmos['L'] == diff['L']]

            if not matching_loads.empty:
                # Pick the one with best Vds headroom or specific Vgs
                # Let's take the one with highest Output Resistance (lowest gds) for Gain
                load = matching_loads.sort_values(by='gds_W').iloc[0]

                W3 = Id_bias_half / load['id_W']

                # OTA Gain = Gm_total / (gds_n + gds_p)
                gds_n = diff['gds_W'] * W1
                gds_p = load['gds_W'] * W3
                gain = target_gm / (gds_n + gds_p)
                gain_db = 20 * np.log10(gain)

                results.append({
                    'L': diff['L'],
                    'Diff_Vgs': diff['Vgs'],
                    'Diff_W': W1,
                    'Load_W': W3,
                    'Load_Vgs': load['Vgs'],
                    'Tail_Current': 2 * Id_bias_half,
                    'Gain_dB': gain_db,
                    'Diff_gm_id': diff['gm_id']
                })

        df_res = pd.DataFrame(results)
        df_res.to_csv(filename, index=False)
        print(f"-> Saved {len(df_res)} valid OTA designs to '{filename}'")
        return df_res

    # ==========================================
    # 3. Miller OTA Solver (Saves ALL valid designs)
    # ==========================================
    def solve_miller_ota(self, target_gbw, cl_load, filename):
        print(f"\n--- Generating Design Space for Miller OTA ---")

        # Cc Calculation
        Cc = 0.3 * cl_load
        gm1_req = target_gbw * 2 * np.pi * Cc
        gm2_req = 10 * gm1_req

        # 1. Solve Stage 1 (Reuse Simple OTA Logic internally)
        # We need a Diff Pair that gives gm1_req
        nmos = self._filter_device("nmos", v_ds_min=0.4, l_min=0.5)

        # 2. Solve Stage 2 (CS Amp PMOS)
        pmos = self._filter_device("pmos", v_ds_min=0.4)

        results = []

        # To avoid massive loops, let's pick optimal Stage 2 candidates first
        # Stage 2 candidates must provide gm2_req with reasonable area
        valid_stg2 = pmos.copy()
        valid_stg2['W2_req'] = gm2_req / valid_stg2['gm_W']
        valid_stg2['Stg2_Current'] = valid_stg2['id_W'] * valid_stg2['W2_req']

        # Filter: Don't allow crazy currents (> 1mA for this example)
        valid_stg2 = valid_stg2[valid_stg2['Stg2_Current'] < 1e-3]

        # Loop through NMOS (Stage 1) and combine with best Stage 2
        for idx, stg1 in nmos.iterrows():
            W1 = gm1_req / stg1['gm_W']
            Itail = 2 * stg1['id_W'] * W1

            # Find a compatible Stage 2 (similar L usually preferred for process tracking)
            match_stg2 = valid_stg2[valid_stg2['L'] == stg1['L']]

            if not match_stg2.empty:
                stg2 = match_stg2.iloc[0] # Pick first valid match

                results.append({
                    'L': stg1['L'],
                    'Cc': Cc,
                    'Stg1_W': W1,
                    'Stg1_Tail': Itail,
                    'Stg2_W': stg2['W2_req'],
                    'Stg2_Current': stg2['Stg2_Current'],
                    'GBW_Target': target_gbw
                })

        df_res = pd.DataFrame(results)
        df_res.to_csv(filename, index=False)
        print(f"-> Saved {len(df_res)} valid Miller designs to '{filename}'")
        return df_res

# --- Execution ---
solver = AnalogSolver("master_lut.csv")

# 1. Generate CSA Designs -> 'csa_results.csv'
solver.solve_single_stage("CSA", target_gain=30, cl_load=1e-12, i_bias_max=10e-6, filename="csa_results.csv")

# 2. Generate OTA Designs -> 'ota_results.csv'
solver.solve_simple_ota(target_gm=100e-6, filename="ota_results.csv")

# 3. Generate Miller Designs -> 'miller_results.csv'
solver.solve_miller_ota(target_gbw=10e6, cl_load=5e-12, filename="miller_results.csv")

print("\nAll files generated successfully.")
```

```python
import subprocess
import os
import sys
import numpy as np
import matplotlib.pyplot as plt

# --- CONFIGURATION ---
SPICE_FILE = "datasheet_sim.spice"
OUTPUT_AC = "data_ac.txt"
OUTPUT_DC = "data_dc.txt"
OUTPUT_TRAN = "data_tran.txt"

def run_datasheet_sim():
    print("--- Generating Full Circuit Datasheet (Miller OTA) ---")

    # 1. Define the Master Netlist with 3 Control Sections
    # We use the generic models to guarantee it runs on your setup.
    netlist_content = f"""
* Miller OTA Master Testbench
.title Miller Datasheet

* --- MODELS ---
.model sky130_fd_pr__nfet_01v8 nmos level=1 vto=0.50 kp=180u lambda=0.03 tox=4.1n
.model sky130_fd_pr__pfet_01v8 pmos level=1 vto=-0.50 kp=80u  lambda=0.03 tox=4.1n

* --- CIRCUIT (Miller OTA) ---
.param w1=10 l1=1 w2=20 l2=1 cc=1.2p itail=10u
.subckt miller_ota vip vin out vdd
    M1 d1 vip tail 0 sky130_fd_pr__nfet_01v8 w={{w1}} l={{l1}}
    M2 d2 vin tail 0 sky130_fd_pr__nfet_01v8 w={{w1}} l={{l1}}
    I_tail tail 0 {{itail}}
    M3 d1 d1 vdd vdd sky130_fd_pr__pfet_01v8 w={{w1}} l={{l1}}
    M4 d2 d1 vdd vdd sky130_fd_pr__pfet_01v8 w={{w1}} l={{l1}}
    M5 out d2 vdd vdd sky130_fd_pr__pfet_01v8 w={{w2}} l={{l2}}
    I_out out 0 10u
    Cc d2 out {{cc}}
.ends

* --- INSTANTIATION ---
Xamp vip vin out vdd miller_ota
Cl out 0 1p
Vdd vdd 0 1.8

* --- SOURCES (Configured for multiple tests) ---
* DC/AC Input
Vin_p vip 0 DC 0.9 AC 1
Vin_n vin 0 DC 0.9
* Pulse Input for Transient (Initially 0, no effect on AC/DC)
Vpulse vip_tran 0 PULSE(0.9 0.91 10n 1n 1n 500n 1u)

* --- CONTROL BLOCK ---
.control
set filetype=ascii

* 1. AC Analysis (Bode Plot)
* --------------------------
ac dec 20 10 100meg
wrdata {OUTPUT_AC} vdb(out) vp(out)

* 2. DC Sweep (Input Common Mode Range / Swing)
* ---------------------------------------------
* We sweep the positive input from 0 to 1.8V to see where gain dies
dc Vin_p 0 1.8 0.01
* Derivative of Vout is the Small Signal DC Gain
wrdata {OUTPUT_DC} v(out)

* 3. Transient Analysis (Step Response)
* -------------------------------------
* Switch input to the Pulse Source
alter @Vin_p[ac] = 0
alter @Vin_p[dc] = 0
* Connect Xamp input to Vpulse (Conceptual rewire)
* Since we can't easily rewire in ngspice batch, we just re-run a separate block or
* use a trick: We used Vin_p for AC/DC. For Tran, we need a step.
* Simpler: We just define the step on Vin_p using 'alter' source function
alter Vin_p pulse(0.9 0.95 100n 10n 10n 2u 4u)
tran 10n 5u
wrdata {OUTPUT_TRAN} v(out) v(vip)

quit
.endc
.end
"""
    with open(SPICE_FILE, "w") as f:
        f.write(netlist_content)

    # 2. Run Simulation
    print("Running Ngspice...")
    subprocess.run(["ngspice", "-b", SPICE_FILE], capture_output=True)
    print("Simulations Complete.")

    # 3. Plotting
    plot_datasheet()

def plot_datasheet():
    fig = plt.figure(figsize=(15, 10))
    plt.style.use('seaborn-v0_8-whitegrid')
    fig.suptitle('Miller OTA Performance Datasheet', fontsize=20, weight='bold')

    # --- Panel 1: AC Response (Bode Plot) ---
    ax1 = fig.add_subplot(2, 2, 1)
    try:
        data = np.loadtxt(OUTPUT_AC)
        freq, gain, phase = data[:, 0], data[:, 1], data[:, 3]

        ax1.semilogx(freq, gain, 'b-', label='Gain (dB)', linewidth=2)
        ax1.set_ylabel('Gain (dB)', color='b', fontsize=12)
        ax1.grid(True, which='both', linestyle='--')

        ax1b = ax1.twinx()
        ax1b.semilogx(freq, phase, 'r--', label='Phase', linewidth=1.5)
        ax1b.set_ylabel('Phase (deg)', color='r', fontsize=12)

        ax1.set_title("1. Frequency Response (AC)", fontsize=14)
        ax1.set_xlabel("Frequency (Hz)")
    except:
        ax1.text(0.5, 0.5, "AC Data Failed", ha='center')

    # --- Panel 2: DC Sweep (Swing) ---
    ax2 = fig.add_subplot(2, 2, 2)
    try:
        data = np.loadtxt(OUTPUT_DC)
        vin, vout = data[:, 0], data[:, 1]

        # Calculate Gain (Derivative) to find linear range
        gain_dc = np.gradient(vout, vin)

        ax2.plot(vin, vout, 'k-', linewidth=2, label='Vout')
        ax2.set_ylabel('Output Voltage (V)', fontsize=12)
        ax2.set_xlabel('Input Voltage (V)', fontsize=12)
        ax2.set_title("2. DC Transfer & Swing", fontsize=14)
        ax2.grid(True)

        # Highlight high gain region
        high_gain_idx = np.where(np.abs(gain_dc) > 10)[0]
        if len(high_gain_idx) > 0:
            ax2.axvspan(vin[high_gain_idx[0]], vin[high_gain_idx[-1]], color='green', alpha=0.1, label='Active Region')
        ax2.legend()
    except:
        ax2.text(0.5, 0.5, "DC Data Failed", ha='center')

    # --- Panel 3: Transient (Step Response) ---
    ax3 = fig.add_subplot(2, 1, 2)
    try:
        data = np.loadtxt(OUTPUT_TRAN)
        time, vout, vin_step = data[:, 0], data[:, 1], data[:, 3]

        ax3.plot(time*1e6, vin_step, 'g--', alpha=0.5, label='Input Step')
        ax3.plot(time*1e6, vout, 'b-', linewidth=2, label='Output Response')

        ax3.set_title("3. Transient Step Response (Stability Check)", fontsize=14)
        ax3.set_xlabel("Time (microseconds)", fontsize=12)
        ax3.set_ylabel("Voltage (V)", fontsize=12)
        ax3.grid(True)
        ax3.legend()

        # Check for ringing
        overshoot = (np.max(vout) - vout[-1]) / (vout[-1] - vout[0])
        if overshoot > 0.1:
            ax3.text(time[len(time)//2]*1e6, np.max(vout), "Ringing Detected!", color='red')
        else:
            ax3.text(time[len(time)//2]*1e6, np.max(vout), "Stable (No Overshoot)", color='green')

    except:
        ax3.text(0.5, 0.5, "Transient Data Failed", ha='center')

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.savefig("full_datasheet.png", dpi=300)
    print("Datasheet saved to 'full_datasheet.png'")
    plt.show()

if __name__ == "__main__":
    run_datasheet_sim()
```

```python
import subprocess
import os
import sys
import numpy as np
import matplotlib.pyplot as plt

# --- CONFIGURATION ---
SPICE_NETLIST = "final_run_v2.spice"
OUTPUT_DATA = "ac_response.txt"
LOG_FILE = "sim_debug.log"

def run_simulation_final():
    print("--- Starting Final Simulation (Debugged) ---")

    # 1. Create the Netlist
    # CHANGES MADE:
    # - Fixed Sources: Combined DC and AC into single lines (Vip, Vin)
    # - Fixed Load: Reduced I_out from 50u to 10u to match Transistor M5 capability
    # - Fixed Prefixes: Ensured all transistors use 'M'

    netlist_content = f"""
* Final Miller OTA Verification (Functional)
.title Miller OTA

* --- GENERIC MODELS ---
.model sky130_fd_pr__nfet_01v8 nmos level=1 vto=0.50 kp=180u lambda=0.03 tox=4.1n
.model sky130_fd_pr__pfet_01v8 pmos level=1 vto=-0.50 kp=80u  lambda=0.03 tox=4.1n

* --- CIRCUIT ---
.param w1=10 l1=1 w2=20 l2=1 cc=1.2p itail=10u

* Diff Pair (NMOS)
M1 d1 vip tail 0 sky130_fd_pr__nfet_01v8 w={{w1}} l={{l1}}
M2 d2 vin tail 0 sky130_fd_pr__nfet_01v8 w={{w1}} l={{l1}}
I_tail tail 0 {{itail}}

* Active Load (PMOS)
M3 d1 d1 vdd vdd sky130_fd_pr__pfet_01v8 w={{w1}} l={{l1}}
M4 d2 d1 vdd vdd sky130_fd_pr__pfet_01v8 w={{w1}} l={{l1}}

* Stage 2 (PMOS)
* M5 is 2x width of M3, so it provides approx 2x current (10uA)
M5 out d2 vdd vdd sky130_fd_pr__pfet_01v8 w={{w2}} l={{l2}}

* Load Current (Matched to M5 capability)
I_out out 0 10u
* Compensation Cap
Cc d2 out {{cc}}

* --- SOURCES (FIXED) ---
Vdd vdd 0 1.8
* Input Positive: DC bias 0.9V AND AC signal 1V
Vip vip 0 DC 0.9 AC 1
* Input Negative: DC bias 0.9V (AC Ground)
Vin vin 0 0.9

* --- CONTROL BLOCK ---
.control
set filetype=ascii
* Run AC Analysis
ac dec 20 10 100meg
* Write output
wrdata {OUTPUT_DATA} vdb(out) vp(out)
quit
.endc

.end
"""
    with open(SPICE_NETLIST, "w") as f:
        f.write(netlist_content)
    print(f"[OK] Netlist '{SPICE_NETLIST}' generated.")

    # 2. Run Ngspice
    print("Running Ngspice...")

    try:
        with open(LOG_FILE, "w") as log:
            result = subprocess.run(
                ["ngspice", "-b", SPICE_NETLIST],
                stdout=log,
                stderr=subprocess.STDOUT,
                text=True
            )

        # 3. Check and Plot
        if os.path.exists(OUTPUT_DATA):
            print(f"[SUCCESS] Data generated in '{OUTPUT_DATA}'.")
            plot_final_data()
        else:
            print("\n[CRITICAL ERROR] Simulation failed.")
            print("--- NGSPICE LOG ---")
            with open(LOG_FILE, "r") as log_read:
                print(log_read.read())

    except FileNotFoundError:
        print("[FAIL] Ngspice not installed.")

def plot_final_data():
    try:
        data = np.loadtxt(OUTPUT_DATA)
        if data.shape[1] >= 4:
            freq, gain, phase = data[:, 0], data[:, 1], data[:, 3]
        else:
            freq, gain, phase = data[:, 0], data[:, 1], data[:, 2]

        fig, ax1 = plt.subplots(figsize=(10, 6))

        # Gain
        color = '#1f77b4'
        ax1.set_xlabel('Frequency (Hz)', fontsize=12)
        ax1.set_ylabel('Gain (dB)', color=color, fontsize=12)
        ax1.semilogx(freq, gain, color=color, linewidth=2.5, label='Gain')
        ax1.tick_params(axis='y', labelcolor=color)
        ax1.grid(True, which='both', linestyle='--', alpha=0.6)

        # Phase
        ax2 = ax1.twinx()
        color = '#ff7f0e'
        ax2.set_ylabel('Phase (deg)', color=color, fontsize=12)
        ax2.semilogx(freq, phase, color=color, linestyle='--', linewidth=2, label='Phase')
        ax2.tick_params(axis='y', labelcolor=color)

        # Markers
        try:
            # Find 0dB crossing
            # If gain is always negative (failed design), this throws error
            idx = np.where(gain < 0)[0][0]
            gbw = freq[idx]
            pm = 180 + phase[idx]

            ax1.plot(gbw, 0, 'o', color='green', markersize=10)

            textstr = f'GBW: {gbw/1e6:.2f} MHz\nPM: {pm:.1f}'
            props = dict(boxstyle='round', facecolor='white', alpha=0.8)
            ax1.text(0.05, 0.15, textstr, transform=ax1.transAxes, fontsize=12,
                    verticalalignment='bottom', bbox=props)

            print(f"\n[FINAL RESULT]\n > Bandwidth: {gbw/1e6:.2f} MHz\n > Phase Margin: {pm:.1f} Degrees")

        except IndexError:
            print("[WARN] Gain is too low (never crossed 0dB). Check biases.")

        plt.title("Verified AC Response: Miller OTA", fontsize=14)
        plt.tight_layout()
        plt.savefig("final_verified_result.png", dpi=300)
        print("[DONE] Image saved to 'final_verified_result.png'")
        plt.show()

    except Exception as e:
        print(f"[ERROR] Plotting failed: {e}")

if __name__ == "__main__":
    run_simulation_final()
```

```python
import subprocess
import os
import sys
import numpy as np
import matplotlib.pyplot as plt

# --- CONFIGURATION ---
SPICE_NETLIST = "verify_simple_ota.spice"
OUTPUT_DATA = "ota_response.txt"
LOG_FILE = "ota_debug.log"

def run_ota_simulation():
    print("--- Starting 5-T OTA Verification ---")

    # 1. Netlist for 5-Transistor OTA
    # This is a Single-Stage Differential Amplifier.
    netlist_content = f"""
* 5-Transistor OTA Verification
.title Simple OTA

* --- GENERIC MODELS ---
.model sky130_fd_pr__nfet_01v8 nmos level=1 vto=0.50 kp=180u lambda=0.03 tox=4.1n
.model sky130_fd_pr__pfet_01v8 pmos level=1 vto=-0.50 kp=80u  lambda=0.03 tox=4.1n

* --- CIRCUIT ---
.param w_diff=10 l_diff=1 w_load=10 l_load=1 itail=20u

* Differential Pair (NMOS)
* M1: Drain=d1, Gate=vip, Source=tail
M1 d1 vip tail 0 sky130_fd_pr__nfet_01v8 w={{w_diff}} l={{l_diff}}
* M2: Drain=out, Gate=vin, Source=tail
M2 out vin tail 0 sky130_fd_pr__nfet_01v8 w={{w_diff}} l={{l_diff}}

* Tail Current Source
I_tail tail 0 {{itail}}

* Active Load (PMOS Current Mirror)
* M3: Diode Connected (Gate connected to Drain d1)
M3 d1 d1 vdd vdd sky130_fd_pr__pfet_01v8 w={{w_load}} l={{l_load}}
* M4: Mirror (Gate connected to d1)
M4 out d1 vdd vdd sky130_fd_pr__pfet_01v8 w={{w_load}} l={{l_load}}

* Load Capacitor
Cl out 0 1p

* --- SOURCES ---
Vdd vdd 0 1.8

* Inputs (Common Mode = 0.9V)
* Vip has the AC signal
Vip vip 0 DC 0.9 AC 1
Vin vin 0 DC 0.9

* --- CONTROL BLOCK ---
.control
set filetype=ascii
* AC Analysis: 10Hz to 100MHz
ac dec 20 10 100meg
wrdata {OUTPUT_DATA} vdb(out) vp(out)
quit
.endc

.end
"""
    with open(SPICE_NETLIST, "w") as f:
        f.write(netlist_content)
    print(f"[OK] OTA Netlist '{SPICE_NETLIST}' generated.")

    # 2. Run Ngspice
    print("Running Ngspice...")
    try:
        with open(LOG_FILE, "w") as log:
            subprocess.run(["ngspice", "-b", SPICE_NETLIST], stdout=log, stderr=subprocess.STDOUT)

        # 3. Plot Results
        if os.path.exists(OUTPUT_DATA):
            plot_ota_data()
        else:
            print("[FAIL] Simulation failed. Check log.")

    except FileNotFoundError:
        print("[FAIL] Ngspice not installed.")

def plot_ota_data():
    try:
        data = np.loadtxt(OUTPUT_DATA)
        if data.shape[1] >= 4:
            freq, gain, phase = data[:, 0], data[:, 1], data[:, 3]
        else:
            freq, gain, phase = data[:, 0], data[:, 1], data[:, 2]

        fig, ax1 = plt.subplots(figsize=(10, 6))

        # Gain
        color = 'green'
        ax1.set_xlabel('Frequency (Hz)')
        ax1.set_ylabel('OTA Gain (dB)', color=color)
        ax1.semilogx(freq, gain, color=color, linewidth=2.5)
        ax1.tick_params(axis='y', labelcolor=color)
        ax1.grid(True, which='both', linestyle='--', alpha=0.6)

        # Phase
        ax2 = ax1.twinx()
        color = 'darkblue'
        ax2.set_ylabel('Phase (deg)', color=color)
        ax2.semilogx(freq, phase, color=color, linestyle='--', linewidth=2)
        ax2.tick_params(axis='y', labelcolor=color)

        # Mark Bandwidth
        try:
            idx = np.where(gain < 0)[0][0]
            gbw = freq[idx]
            pm = 180 + phase[idx]

            ax1.plot(gbw, 0, 'rx', markersize=12)
            ax1.text(gbw, 5, f" GBW: {gbw/1e6:.2f} MHz\n PM: {pm:.1f}", color='red')
            print(f"\n[OTA RESULT] GBW: {gbw/1e6:.2f} MHz | PM: {pm:.1f} Deg")
        except:
            pass

        plt.title("5-Transistor OTA AC Response")
        plt.tight_layout()
        plt.savefig("ota_result.png")
        print("[DONE] Plot saved as 'ota_result.png'")
        plt.show()

    except Exception as e:
        print(f"[ERROR] Plotting failed: {e}")

if __name__ == "__main__":
    run_ota_simulation()
```
